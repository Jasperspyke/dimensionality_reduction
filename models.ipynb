```python
import matplotlib.pyplot as plt
import torch
```

```python
torch.set_default_dtype(torch.float32)
identity = torch.eye(3)
mu = torch.tensor([0,0,0],dtype=torch.float32)
sigma = torch.tensor([1,2,0],dtype=torch.float32)
sigma = torch.eye(3)
size = torch.tensor(30)
second_size = torch.tensor(10)
second_mu = torch.tensor([5,12,0],dtype=torch.float32)
second_sigma = torch.tensor([3,8,2],dtype=torch.float32)
```

```python
cov1 = torch.mul(identity,sigma)
cov2 = torch.mul(identity,second_sigma)
```

```python
class_1 = torch.distributions.MultivariateNormal(loc=mu, covariance_matrix=cov1)
class_2 = torch.distributions.MultivariateNormal(loc=second_mu, covariance_matrix=cov2)
```

```python
sample = class_1.rsample((64, 1))
sample = sample.squeeze()
other_sample = class_2.rsample((64, 1))
other_sample = other_sample.squeeze()
```

```python

y_axis = torch.linspace(0,64,100)
plt.title("Scatter plot of normally distributed data")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.xlim(left=-12, right=12)
plt.ylim(bottom=-12, top=15)
plt.grid(True)
plt.scatter(sample[:,0], sample[:,1])
plt.scatter(other_sample[:,0], other_sample[:,1])
```
